```python
import pandas as pd
from pandas import read_csv
import pandas as pd
import matplotlib.pyplot as plt
import math
  
import matplotlib.pyplot as plt
import numpy as np
import scipy
  
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

import seaborn as sns

df = read_csv('vocoded.csv')
```

```python
regression_model = LinearRegression()
m, b = np.polyfit(df.Distance, df.Gmean, 1)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax2 = ax1.twiny()


ax1.plot(df.Distance, df.Gmean, 'x')
ax1.set_xlabel(r"Actual Positions on Slider")
ax1.set_ylabel(r"Geometric Means of Estimated Position")
ax1.plot(df.Distance, m*df.Distance + b, label = 'Estimated')
ax1.plot(df.Distance, df.Distance, label = 'Ideal')
plt.grid(axis='both')


ax1.legend()

ax2.plot(df.bpm, df.Gmean, 'x')

ax2.set_xlabel(r"BPM (Beats per Minute)")
plt.text(10,12, 'm=0.853', fontsize=10)
plt.text (10,11,'std_err= 0.0342 $r^2$=0.98',fontsize=10)
plt.title("Vocoded Beating Condition",color = 'blue')
plt.savefig("vocoded_graph.png", dpi=400, bbox_inches='tight')
plt.show()
```
